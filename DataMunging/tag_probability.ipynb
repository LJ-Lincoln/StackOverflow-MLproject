```python
import numpy as np
import pandas as pd
import re
from nltk import word_tokenize, FreqDist
```

```python
post = pd.read_pickle('/Users/boyaliu/Documents/UCD/STA 208/final project/qpost_df_new.pkl')
```

```python
tag = post.Tags.map(lambda i: re.sub('<|>', ' ', i))
```

```python
tags =tag.map(lambda i:  " ".join([word for word in word_tokenize(i)]))
```

```python
# create a list contain all the tags
T = tags.tolist()
tag_l = []
for i in T:
    for j in i.split():
        tag_l.append(j)
```

```python
tag_l
```

```python
fdist = FreqDist(tag_l)
```

```python
fdist
```

```python
def tag_prob(i):
    tag_sum = []
    for j in i.split():
        tag_freq = fdist.freq(j)
        tag_sum.append(tag_freq)
    return(sum(tag_sum))
        
```

```python
tag_probability = tags.map(lambda i: tag_prob(i))
```

```python
len(tag_probability)
```

```python
import pickle
with open('tag_probability.pickle', 'wb') as handle:
    pickle.dump(tag_probability, handle,protocol=2)
```
