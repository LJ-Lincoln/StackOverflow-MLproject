```python
import numpy as np
import pandas as pd
import re
from nltk import word_tokenize, FreqDist
```

```python
post = pd.read_pickle('/Users/boyaliu/Documents/UCD/STA 208/final project/qpost_df_new.pkl')
```

```python
tag = post.Tags.map(lambda i: re.sub('<|>', ' ', i))
```

```python
tags =tag.map(lambda i:  " ".join([word for word in word_tokenize(i)]))
```

```python
tags[0:4]
```

```python
# create a list contain all the tags
T = tags.tolist()
tag_l = []
for i in T:
    for j in i.split():
        tag_l.append(j)
```

```python
tag_l
```

```python
fdist = FreqDist(tag_l)
```

```python
fdist
```

```python
# calculate the probability for each tag
def tag_prob(i):
    tag_sum = []
    for j in i.split():
        tag_freq = fdist.freq(j)
        tag_sum.append(tag_freq)
    return(sum(tag_sum))
        
```

```python
tag_probability = tags.map(lambda i: tag_prob(i))
```

```python
tag_probability.quantile([.2,.4,.6,.8])
```

```python
def tag_rank(i):
    if i >= 0.058459:
        rank = 0
    elif i >= 0.025726:
        rank = 1
    elif i >= 0.013992:
        rank = 2
    elif i >= 0.005903:
        rank = 3
    else:
        rank = 4
    return(rank)        
```

```python
tag_rank = tag_probability.map(lambda i: tag_rank(i))
```
